```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('padaria.csv', encoding='latin-1', sep=';')


df.columns = df.columns.str.strip().str.lower()


coluna_produto = None
for col in df.columns:
    if 'prod' in col:
        coluna_produto = col
        break

if coluna_produto is None:
    coluna_produto = df.columns[0]


df[coluna_produto].value_counts().head(5).plot(
    kind='pie',
    autopct='%1.1f%%'
)

plt.title('Top 5 produtos mais vendidos')
plt.ylabel('')  # tira "y"
plt.show()
```
